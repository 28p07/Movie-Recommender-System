### Importing Libraries

```python
import numpy as np
import pandas as pd
import ast
```

### Loading dataset

```python
movies  = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.info()
```

All the columns in the dataset are not relevant for recommendation puprose.We choose only a few which are:
1. id
2. title
3. genre
4. keywords
5. overview
6. cast
7. crew

```python
movies = movies[['id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

### Preprocessing

```python
g = ast.literal_eval(movies['genres'][0])
g
```

```python
def convert(g):
    l = []
    g = ast.literal_eval(g)
    for i in g:
        l.append(i['name'])
    return l    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert(g):
    l = []
    g = ast.literal_eval(g)
    counter = 0
    for i in g:
        if counter<3:
            l.append(i['name'])
        counter+=1
    return l 
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def convert_crew(g):
    l = []
    g = ast.literal_eval(g)
    for i in g:
        if i['job'] == 'Director':
            l.append(i['name'])
    return l  
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
def remove_space(g):
    g1 = []
    for i in g:
        g1.append(i.replace(" ",""))
    return g1
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
df = movies[['id','title','tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

### Stemming

```python
df.head()
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
df['tags'] = df['tags'].apply(lambda x:ps.stem(x))
```

```python
df.head()
```

### Text Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
cv = vectorizer.fit_transform(df['tags']).toarray()
```

```python
cv.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(cv)
```

```python
similarity.shape
```

```python
df.head()
```

```python
sorted(list(enumerate(similarity[0])),reverse =True,key=lambda x:x[1])
```

### Final Results

```python
def recommend(movie):
    index = df[df['title'] == movie].index.item()
    recommendations = sorted(enumerate(similarity[index]),reverse =True,key=lambda x:x[1])[1:6]
    for i in recommendations:
        print(i[0])
```

```python
recommend('Avatar')
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    recommendations = sorted(enumerate(similarity[index]),reverse=True,key=lambda x:x[1])[1:6]
    for i in recommendations:
        print(df['title'][i[0]])
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
